```python
import pandas as pd
import numpy as np
import json
import math
```

```python
df = pd.read_json('debate_data.json')
df
```

```python
df_god = df[df['Target']=='god']
nu = math.ceil((30 * len(df_god)) / 100)
df_god = df_god.sample(n= nu)
df_god
```

```python
df_creation = df[df['Target']=='creation']
nu = math.ceil((30 * len(df_creation)) / 100)
df_creation = df_creation.sample(n= nu)
df_creation
```

```python
df_abortion = df[df['Target']=='abortion']
nu = math.ceil((30 * len(df_abortion)) / 100)
df_abortion = df_abortion.sample(n= nu)
df_abortion
```

```python
df_gayRights = df[df['Target']=='gayRights']
nu = math.ceil((30 * len(df_gayRights)) / 100)
df_gayRights = df_gayRights.sample(n= nu)
df_gayRights
```

```python
df_guns = df[df['Target']=='guns']
nu = math.ceil((30 * len(df_guns)) / 100)
df_guns = df_guns.sample(n= nu)
df_guns
```

```python

df_healthcare = df[df['Target']=='healthcare']
nu = math.ceil((30 * len(df_healthcare)) / 100)
df_healthcare = df_healthcare.sample(n= nu)
df_healthcare
```

```python
df_test = pd.concat([df_abortion, df_creation, df_gayRights, df_god, df_guns, df_healthcare])
df_test
```
