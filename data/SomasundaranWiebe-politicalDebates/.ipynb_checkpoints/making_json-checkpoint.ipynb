```python
import pandas as pd
import numpy as np
import os
import glob
import re
```

```python
path = '/home/parush/Desktop/Stance/Dataset/The Political Debate Corpus/political_debates_SomasundaranWiebeAcl2009/SomasundaranWiebe-politicalDebates'
target = ['god','gayRights','guns','abortion','healthcare','creation']
```

```python
def remove_URL(sample):
    """Remove URLs from a sample string"""
    return re.sub(r"http\S+", "", sample)
```

```python
c = 0
data = {'Tweet':[],'Target':[],'Stance':[]}
for t in target:
    files = glob.glob(path + '/' + t + '/*',  recursive = True)
    for file in files:
        c += 1
        
        with open(file, 'r', encoding = "ISO-8859-1") as f:
            content = []
            for line in f:
                line.rstrip()
                content.append(line)
            data['Tweet'].append(remove_URL(content[3]))
            data['Target'].append(t)
            data['Stance'].append('FAVOR' if content[0][8:15] == 'stance1' else 'AGAINST')
            content.clear()
                
                
                
        
    
    
    
```

```python
df = pd.DataFrame(data)
np.unique(df['Target'])
```

```python
import json
with open("debate_data.json", "w") as outfile: 
    json.dump(data, outfile)
```
